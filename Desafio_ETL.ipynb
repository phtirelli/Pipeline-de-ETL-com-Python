EXTRACT


```python
import pandas as pd #importando biblioteca pandas


users = pd.read_csv('userDIO.csv').to_dict(orient='records') # lendo o arquivo e salvando em um novo dicionario

for user in users: #percorrendo o dicionario e adicionando a coluna news
    user['news'] = []

print(users)
```

TRANSFORM

```python
for user in users: #percorrendo o dicionario e adicionando uma imagem e mensagem de feliz ano novo

    user['news'] = [
        {
            "icon": "https://cdn.pensador.com/img/imagens/fe/li/feliz_ano_novo_que_venham_novos_dias_de_muito_sucesso_e_felicidade_1.jpg",
            "description": "Feliz Ano Novo"
        }
    ]

print(users[0]) #imprimindo o primeiro usuario para verificação
```

LOAD

```python
df_updated = pd.DataFrame(users) # converte o dicionario em um dataframe
df_updated.to_csv('users_updated.csv', index=False) #salvando o dataframe em um novo arquivo csv
```
